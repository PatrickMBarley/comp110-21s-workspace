Step 8: Create a mask that filters 

```python
def no_experience(col: list[str]) -> list[bool]:
    result: list[bool] = []
    for item in col:
        if item == "None to less than one month!":
            result.append(True)
        else:
            result.append(False)
    return result

experience_mask: list[bool] = no_experience(selected_data["prior_exp"])
print(experience_mask)
```

Step 9: This uses the mask

```python
def masked(col: list[str], mask: list[bool]) -> list[str]:
    result: list[str] = []
    for i in range(len(mask)):
        if mask[i]:
            result.append(col[i])
    return result

def not_masked(col: list[str], mask: list[bool]) -> list[str]:
    result: list[str] = []
    for i in range(len(mask)):
        if not mask[i]:
            result.append(col[i])
    return result   

unmasked = masked(selected_data["difficulty"], experience_mask)
not_unmasked = not_masked(selected_data["difficulty"], experience_mask)
print(not_unmasked)
```

Step 10: Create data visualizations of

```python
from matplotlib import pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)
fig.suptitle("Difficulty Ratings in COMP 110 - Spring 2021")

unmasked_counts: dict[str, list[str]] = count(unmasked)
not_unmasked_counts: dict[str, list[str]] = count(not_unmasked)

axes[0].set_title("Students With No Experience")
axes[0].bar(unmasked_counts.keys(), unmasked_counts.values())
axes[0].tick_params(axis='x', labelrotation = 45)

axes[1].set_title("Students With Experience")
axes[1].bar(not_unmasked_counts.keys(), not_unmasked_counts.values())
axes[1].tick_params(axis='x', labelrotation = 45)
```
